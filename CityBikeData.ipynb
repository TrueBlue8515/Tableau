```python
import pandas as pd
import numpy as np
```

```python
#Read in data from csv files
df1 = pd.read_csv(r"C:\Users\halcm\Desktop\Tableau\Data\201306_citibike_tripdata.csv")
df2 = pd.read_csv(r"C:\Users\halcm\Desktop\Tableau\Data\201406_citibike_tripdata.csv")
df3 = pd.read_csv(r"C:\Users\halcm\Desktop\Tableau\Data\201506_citibike_tripdata.csv")
df4 = pd.read_csv(r"C:\Users\halcm\Desktop\Tableau\Data\201606_citibike_tripdata.csv")
df5 = pd.read_csv(r"C:\Users\halcm\Desktop\Tableau\Data\201706_citibike_tripdata.csv")
#df6 = pd.read_csv(r"C:\Users\halcm\Desktop\Tableau\Data\201806_citibike_tripdata.csv")

#Take a glimpse at the data
df1.head()
```

```python
#Append remaining dataframes to initial dataframe
df1 = df1.append([df2, df3, df4, df5])
```

```python
#Determine how many rows I'm working with for each column
df1.count()
```

```python
#Remove missing values
df1 = df1.dropna(how='any')

#See if removing the missing values evened out the data
df1.count()
```

```python
#CODE NOT USED - Options considered when trying to remove '\N' as value in birth year column
#df1 = df1.replace({'birth year' : '\\N'}, value='NaN')

#df1.replace(to_replace=r"^\N.$"", value='NaN', regex=True)

#df1 = df1[df1['birth year'] != 1800]
```

```python
#Rename columns due to Tableau not liking column headers to be more than 20 characters
df1=df1.rename(columns = {
    'tripduration':'Trip Duration',
    'starttime':'Start Time',
    'stoptime':'Stop Time',
    'start station id':'Start Station ID',
    'start station name':'Start Station Name',
    'start station latitude':'Start Station Lat',
    'start station longitude':'Start Station Long',
    'end station id':'End Station ID',
    'end station name':'End Station Name',
    'end station latitude':'End Station Lat',
    'end station longitude':'End Station Long',
    'bikeid':'Bike ID',
    'usertype':'User Type',
    'birth year':'Birth Year',
    'gender':'Gender'
})
```

```python
#Separate the Date from the Start Time & Stop Time columns

# new data frame with split value columns 
new_df1 = df1["Start Time"].str.split(" ", n = 1, expand = True)
#df1 = df1["stoptime"].str.split(" ", n = 1, expand = True)
  
# making seperate date from new data frame 
df1["Date"]= new_df1[0] 
  
# making seperate start time column from new data frame 
df1["Start Time"]= new_df1[1]
#df1["Stop Time"]= new_df1[1]
  
# Dropping old columns 
df1.drop(columns =["Start Time", "Stop Time"], inplace = True)
#df1.drop(columns =["stoptime"], inplace = True)

df1.head()
```

```python
#Reorder the columns in the dataframe
df1 = df1[[
    'Date',
    'Trip Duration',
    'Start Station ID',
    'Start Station Name',
    'Start Station Lat',
    'Start Station Long',
    'End Station ID',
    'End Station Name',
    'End Station Lat',
    'End Station Long',
    'Bike ID',
    'User Type',
    'Birth Year',
    'Gender'
]]
df1.head()
```

```python
#Final check to make sure nothing has changed
df1.count()
```

```python
#Save appended and cleaned dataframe to a new csv
df1.to_csv(r"C:\Users\halcm\Desktop\Tableau\Data\Clean_Data.csv")
```
